```python
# Importing Necessary Libraries
# Import required libraries for time, csv, and random operations.
import time, csv, random

# Signup Function
# This function handles user signup and login processes.
# Depending on whether the user already has an account, it either logs them in or registers a new account.
def Signup(v):
    if v == "Y":
        time.sleep(1)
        print('please login to continue')
        time.sleep(2)

        user_id = input("please enter your email id:  ")
        time.sleep(1)
        password = input("please enter your password:  ")
        time.sleep(1)
        print('Logging in...')
        time.sleep(1)
    elif v == "N":
        time.sleep(1)
        print('please register to continue')

        time.sleep(1)
        FN = input('Enter your Full Name\n')
        CN = input('Enter your Contact Number\n')
        Email = input('Enter your Email\n')
        Address = input('Enter your Address\n')
        time.sleep(1)

        print('Successfully Registered! Please Refresh and login to continue\n')
        time.sleep(1)
        user_id = input("please enter your email id:  ")
        time.sleep(1)
        password = input("please enter your password:  ")
        time.sleep(1)
        print('Logging in...')
        time.sleep(1)
```

```python
# S_action Function
# This function handles various product management actions based on user input.
# Actions include adding a new product, updating an existing product, viewing product details, and deleting a product.

def S_action(a):
    import time, csv

    if a == 1:
        # Adding a New Product
        # This block allows the user to add a new product by entering its details.
        time.sleep(1)
        print("please enter the details of the product you want to add")
        productname = input('enter product name:  ')
        time.sleep(1)
        mfname = input('enter manufacturer name:  ')
        time.sleep(1)
        price = input('enter price:  ')
        time.sleep(1)
        quantity = input('enter quantity:  ')
        time.sleep(2)

        print("Congrats! Successfully added the product")

        with open(project_dir+'Product_details.csv', 'a') as f:
            file = csv.writer(f)
            file.writerow(['PROD', productname, mfname, price, quantity])

    elif a == 2:
        # Updating an Existing Product
        # This block allows the user to update details of an existing product by its ID.
        UP2 = {}
        time.sleep(1)

        print("please enter the ID of the product you want to update")
        prod = input('Product ID:  ')

        time.sleep(1)
        print("press 2 to change Manufacturer")
        print("press 3 to change Price")
        print("press 4 to change Quantity")
        time.sleep(1)

        I = input('What property do you want to change/update? ')
        U = input('Please write the updated entry ')

        UP2[str(I)] = I
        product_found = False
        with open(project_dir + 'Product_details.csv', 'r') as in_file:
            reader = list(csv.reader(in_file))
        for row in reader:
            if row[0] == prod:
                product_found = True  # Set the flag to True when product is found
                print('Updating row for PROD_ID ', row[0], " and ", row[1])
                row[int(I)] = U
                break  # Exit the loop once the product is found
        if product_found:
            with open(project_dir + 'Product_details.csv', 'w', newline='') as out_file:
                writer = csv.writer(out_file)
                writer.writerows(reader)  # Write all rows back to the CSV file
            time.sleep(2)
            print('Update completed')
        else:
            print("Product not found")

    elif a == 3:
        # Viewing a Product
        # This block allows the user to view details of a product by its ID.
        time.sleep(1)

        print("please enter the details of the product you want to view")
        prod = input('Product ID:  ')
        product_found = False

        with open(project_dir+'Product_details.csv', 'r') as f:
            file = csv.reader(f)
            for row in file:
                if row and row[0] == str(prod):
                    product_found = True
                    print("Below are the details of the product you are looking for: \n")
                    time.sleep(1)
                    print("Product name:  ", row[1])
                    print("Product manufacturer:  ", row[2])
                    print("Product Price:  ", row[3])
                    print("Product Quantity:  ", row[4])
                    break
            if not product_found:
                print("Product not found")

    elif a == 4:
        # Deleting a Product
        # This block allows the user to delete a product by its ID.
        time.sleep(1)

        print("please enter the details of the product you want to delete")
        prod = input('Product ID:  ')
        product_found = False  # Flag to check if the product is found
        with open(project_dir + 'Product_details.csv', 'r') as in_file:
            reader = list(csv.reader(in_file))
        new_rows = []
        for row in reader:
            if row[0] != prod:
                new_rows.append(row)
            else:
                product_found = True
        if product_found:
            with open(project_dir + 'Product_details.csv', 'w', newline='') as out_file:
                writer = csv.writer(out_file)
                writer.writerows(new_rows)  # Write non-deleted rows back to the CSV file
            time.sleep(1)
            print("You have successfully removed the product.")
        else:
            print("Product not found")
    else:
        print('expecting better input')
```

```python
# Product Management Options
# This function presents the user with different actions they can perform on products.
# Actions include creating, updating, viewing, and deleting products.
def option():
    import time
    print('press 1 to Create Product')
    time.sleep(1)
    print('press 2 to Update Product')
    time.sleep(1)
    print('press 3 to View Product')
    time.sleep(1)
    print('press 4 to Delete Product')
    action = int(input('your input: '))
    return action
```

```python
# Main Execution Block
# This block initiates the seller application.
# It welcomes the user, checks if they have an account, and proceeds based on their input.
print("Welcome to seller application \n")
print('Hello there! ')
user = input("Do you already have an account?")

# Execute Signup and Product Management Actions
Signup(user)
S_action(option())
# Sample product ID: D1PROD01
```
